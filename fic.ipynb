```python
import pandas as pd
import matplotlib.pyplot as mpl
import seaborn as sns

pl = pd.read_csv('EPL_20_21.csv')

pl = pl.assign(sum_g_a = 0)
pl['sum_g_a'] = pl['Goals'] + pl['Assists']
```

```python
pl.head(10)
```

```python
#Impacte du nombre de minute jouer sur les performance des joueeur 
sns.regplot(x = pl['Mins'], y= pl['sum_g_a'], fit_reg=True, order=4)
```

```python
#impact de l'age sur le nombre de match joue
sns.displot(x=pl['Mins'], y=pl['sum_g_a'])
```

```python
sns.displot(data=pl, x="Mins",kde=True)
```

```python
sns.catplot(data=pl, x='sum_g_a', y='Mins', kind='bar', height=8.27, aspect=11.7/8.27).set_xticklabels(rotation=90)
```

```python
sns.catplot(data=pl, x='Position', y='sum_g_a', kind='bar', height=8.27, aspect=11.7/8.27).set_xticklabels(rotation=90)
```

```python
sns.displot(data=pl, x="sum_g_a",kde=True)
```
